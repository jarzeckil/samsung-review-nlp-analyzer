```python
import dateparser
import pandas as pd
import seaborn as sns
from src.core.config import DATA_DIR
```

```python
df = pd.read_csv(DATA_DIR / 'reviews_uk.csv', usecols=[1, 2, 3])
```

```python
df
```

```python
df = df.dropna(how='any')
```

```python
df['Date'] = df['Date'].apply(lambda x: dateparser.parse(x, languages=['en']))
```

```python
df.head()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
sns.histplot(df['Score'])
```

```python
sns.boxplot(x=df['Score'], y=df['Review'].apply(lambda x: len(x)))
```

```python
years = df.copy()
years['year'] = years['Date'].apply(lambda x: x.year)
years.groupby('year').count()
```

```python
df.to_csv(DATA_DIR / 'reviews_uk_cleaned.csv', index=False)
```
