```python
import dateparser
import pandas as pd
import seaborn as sns
from src.core.config import DATA_DIR
```

```python
df = pd.read_csv(DATA_DIR / 'reviews.csv', usecols=[1, 2, 3])
```

```python
df
```

Dropping rows without review

```python
df = df.dropna(how='any')
```

Parsing date to standard format

```python
df['Date'] = df['Date'].apply(lambda x: dateparser.parse(x, languages=['en']))
```

```python
df.head()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

Most scores are low - people that are satisfied generally don't care to leave a review

```python
sns.histplot(df['Score'])
```

Worst scores have longest reviews - same case as before - if people are satisfied at most they'll leave 5 stars and go do sth else

```python
sns.boxplot(x=df['Score'], y=df['Review'].apply(lambda x: len(x)))
```

Most reviews are new (scraping's fault)

```python
years = df.copy()
years['year'] = years['Date'].apply(lambda x: x.year)
years.groupby('year').count()
```

```python
df.to_csv(DATA_DIR / 'reviews_cleaned.csv', index=False)
```
